```python
## Check Functions ##

def CheckForDtype(dataframe):
  df = dataframe
  dtypeSeries = df.dtypes
  colnameSeries = df.columns.values
  index = 0
  dtype_issue = 0
  for a in dtypeSeries:
    if(a == "object"):
      dtype_issue = 1
    else:
      index = index + 1
  return dtype_issue

def CheckForMissing(dataframe):
  df = dataframe
  isNullSeries = df.isnull().any(axis=1)
  for a in isNullSeries:
    if(a == True):
      null = True
      break
    else:
      null = False
  return null

def CheckForDuplicated(dataframe):
  df = dataframe
  isDuplicatedSeries = df.duplicated()
  for a in isDuplicatedSeries:
    if(a == True):
      duplicate = True
      break
    else:
      duplicate = False
  return duplicate

def CheckForIdColumn(dataframe):
  df = dataframe
  col_names = df.columns.values
  index = 0
  for a in col_names:
    if(a == 'ID' or a == 'id' or a == "Id" or a == 'iD'):
      id = True
      break
    elif(a.includes("_ID") or a.includes('_id') or a.includes('_Id') or a.includes('_iD')):
      id = True
      break
    else:
      id = False
      index = index + 1
  return index

def CheckForCompleteIncrementalColumn(dataframe,index):
  df = dataframe

  row_count = (len(df)) + 1
  expected_row_id = set(range(1,row_count))
  actual_row_id = set(df.columns[index])
  is_incremental = expected_row_id == actual_row_id

  if(is_incremental == True):
    increment = 1
  else:
    increment = 0

  return increment

def CheckDataAlignment(dataframe):
  df = dataframe
  alignment = "WIP"
  # Check if dataframe is horizontal or vertically aligned
  return alignment

def CheckForCompleteUniqueColumn(dataframe,index):
  df = dataframe
  col_name = df.columns[index]
  row_id = pd.Series(df[col_name].map(type))
  unique_row_id = pd.Series(row_id.unique())
  n_row_id = len(row_id)
  n_urow_id = len(unique_row_id)

  if(n_row_id == n_urow_id):
    completelyUnique = 1
  else:
    completelyUnique = 0
  return completelyUnique

```

```python
## Main Solve Functions ##

def MakeIdColumn(dataframe):
  df = dataframe

  df = df.reset_index()
  df = df.rename(columns = {'index':'row_id'})

  return df

def HandleMissingData(dataframe):
  df = dataframe
  return df

def NoteMissingData(dataframe):
  df = dataframe
  Note = 0
  return Note

def NoteDuplicatedData(dataframe):
  df = dataframe
  Note = 0
  return Note

def NoteColumnMinorities(dataframe):
  df = dataframe
  Note = 0
  return Note

def HandleDuplicatedData(dataframe):
  df = dataframe
  return df

def ConformAllColumnsToMajority(dataframe):
  df = dataframe
  dtypeSeries = df.dtypes
  colnameSeries = df.columns.values
  index = 0
  for a in dtypeSeries:
    if(a == "object"):
      col_name = colnameSeries[index]
      df_dtypes = pd.Series(df[col_name].map(type))
      df_dtypes_unique = pd.Series(df_dtypes.unique())
      string_series = pd.Series(df_dtypes, dtype="string")
      df_dtypes_sorted = string_series.sort_values()
      conv_df_dtypes_unique = df_dtypes_unique.astype(str)
      conv_df_dtypes = df_dtypes.astype(str)
      conv_df_dtypes_sorted = df_dtypes_sorted.astype(str)

      datatypes = config_list[0]
      counter = [0] * datatypes

      counter = CountRowsPerDatatype(counter, conv_df_dtypes_sorted, conv_df_dtypes_unique)
      majority_value = GetMajorityValue(counter, datatypes, conv_df_dtypes_unique)

      if(majority_value == "<class 'str'>"):
        df[col_name] = df[col_name].astype("string")

      elif(majority_value == "<class 'int'>"):
        df[col_name] = df[col_name].astype("Int64")

      elif(majority_value == "<class 'float'>"):
        df[col_name] = df[col_name].astype("Float64")

      elif(majority_value == "<class 'complex'>"):
        df[col_name] = df[col_name].astype(complex)

      elif(majority_value == "<class 'bool'>"):
        df[col_name] = df[col_name].astype("boolean")

    else:
      index = index + 1

  return df
```

```python
## Auxillary Functions ##

def RenameIdColumn(dataframe,index):
  df = dataframe
  df = df.rename(columns={df.columns[index] : "row_id"})

  return df

def CountRowsPerDatatype(array, data_list, data_list_unique):

  counter = array

  for b in data_list:
        x = 0
        for c in data_list_unique:
          if(b == c):
            counter[x] = counter[x] + 1
            break
          else:
            x = x + 1

  return counter

def GetMajorityValue(array,datatypes,data_list_unique):

  counter = array
  conv_df_dtypes_unique = data_list_unique

  new_counter = counter.copy()
  new_counter.sort()
  majority_amount = new_counter[-1]

  m = 0
  for d in counter:
    if(counter[m] == majority_amount and m<datatypes):
      majority_index = m
      break
    else:
      m = m + 1

  majority_value = conv_df_dtypes_unique[majority_index]
  return majority_value


def PrintCounter(array):

  x = 0
  counter = array

  for d in counter:
    print(counter[x])
    x = x + 1
```
